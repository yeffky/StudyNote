```python
import numpy as np
```

# 创建数组

```python
print(np.eye(5))
print(np.array([1, 2, 3, 5]))
print(np.zeros(5))
print(np.ones(5))
print(np.arange(1, 10, 2))
print(np.linspace(18, 20, 12))
print(np.eye(5).shape)
```

# 数组属性

```python
a = np.random.rand(2, 3)
print(a.shape, a.dtype, a.size, a.ndim)
```

# 数学操作

```python
print(a)
print(np.sum(a))
print(np.std(a))
print(np.max(a))
print(np.mean(a))
```

# 随机数生成

```python
# 指定维度，样本位于[0, 1)的随机样本 
print(np.random.rand(2, 3))
# 指定维度，样本为正态分布随机值的数组
print(np.random.randn(2, 3))
```
